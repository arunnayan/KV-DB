```python
import numpy as np

# Example function: f(x) = x^2
def f(x):
    return x**2

# Derivative of the function: f'(x) = 2x
def gradient_f(x):
    return 2*x

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    for _ in range(num_iterations):
        grad = gradient_f(x)
        x = x - learning_rate * grad
        print(f"Iteration {_+1}: x = {x}, f(x) = {f(x)}")
    return x

# Parameters
starting_point = 10  # Initial guess
learning_rate = 0.1
num_iterations = 100

# Finding the minimum
min_x = gradient_descent(starting_point, learning_rate, num_iterations)
print(f"Found minimum at x = {min_x}, f(x) = {f(min_x)}")
```

```python
# Example function: f(x) = -x^2 (since we want to maximize -x^2, which is equivalent to finding the minima of x^2)
def f_neg(x):
    return -x**2

# Derivative of the function: f'(x) = -2x
def gradient_f_neg(x):
    return -2*x

# Gradient Ascent Algorithm
def gradient_ascent(starting_point, learning_rate, num_iterations):
    x = starting_point
    for _ in range(num_iterations):
        grad = gradient_f_neg(x)
        x = x + learning_rate * grad  # Notice the addition instead of subtraction
        print(f"Iteration {_+1}: x = {x}, f(x) = {f_neg(x)}")
    return x

# Parameters
starting_point = 10  # Initial guess
learning_rate = 0.1
num_iterations = 100

# Finding the maximum
max_x = gradient_ascent(starting_point, learning_rate, num_iterations)
print(f"Found maximum at x = {max_x}, f(x) = {f_neg(max_x)}")
```

```python
import numpy as np

# Define the function
def f(x, y):
    return x**2 - y**2

# Define the gradient of the function
def gradient_f(x, y):
    return np.array([2*x, -2*y])

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    point = np.array(starting_point, dtype=float)
    for i in range(num_iterations):
        grad = gradient_f(point[0], point[1])
        point -= learning_rate * grad
        print(f"Iteration {i+1}: point = {point}, f(point) = {f(point[0], point[1])}")
    return point

# Parameters
starting_point = [1.0, 1.0]  # Initial guess
learning_rate = 0.1
num_iterations = 100

# Finding the saddle point
saddle_point = gradient_descent(starting_point, learning_rate, num_iterations)
print(f"Found saddle point at point = {saddle_point}, f(point) = {f(saddle_point[0], saddle_point[1])}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return x**2 - y**2

# Define the gradient of the function
def gradient_f(x, y):
    return np.array([2*x, -2*y])

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, num_iterations):
    points = [starting_point]  # Store the points to plot the path
    point = np.array(starting_point, dtype=float)
    for i in range(num_iterations):
        grad = gradient_f(point[0], point[1])
        point -= learning_rate * grad
        points.append(point.copy())
        if np.linalg.norm(grad) < 1e-5:  # Convergence condition
            break
    return np.array(points)

# Parameters
starting_point = [1.5, 1.5]  # Initial guess
learning_rate = 0.1
num_iterations = 100

# Run gradient descent
points = gradient_descent(starting_point, learning_rate, num_iterations)

# Generate data for plotting
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the function surface and contour with gradient descent path
fig = plt.figure(figsize=(12, 6))

# 3D Surface plot
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.plot(points[:, 0], points[:, 1], f(points[:, 0], points[:, 1]), color='red', marker='o')
ax.set_title('3D Surface Plot of f(x, y) = x^2 - y^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

# Contour plot with gradient descent path
plt.subplot(1, 2, 2)
plt.contour(X, Y, Z, levels=30, cmap='viridis')
plt.quiver(X, Y, gradient_f(X, Y)[0], gradient_f(X, Y)[1], color='red', alpha=0.5)
plt.plot(points[:, 0], points[:, 1], color='red', marker='o')
plt.scatter(points[-1, 0], points[-1, 1], color='blue', label='Saddle Point')  # Saddle point
plt.title('Contour Plot with Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```
