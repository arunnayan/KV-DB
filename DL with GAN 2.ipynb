### **2. `from tensorflow.keras.models import Sequential`**

#### **Theory**

- **`Sequential` Model**: A type of model in Keras, which is a high-level API within TensorFlow designed for easy model building and training.

- **Purpose**: The `Sequential` class provides a straightforward way to construct neural network models where layers are added in a sequential manner, one after another.

- **Concepts:**
  - **Linear Stack of Layers**: The `Sequential` model allows you to create a neural network where each layer has a single input and output. Layers are added in a sequence, with the output of one layer feeding into the next.

### **3. `from tensorflow.keras.layers import Dense`**

#### **Theory**

- **`Dense` Layer**: A fundamental type of layer used in neural networks, where every neuron in the layer is connected to every neuron in the previous layer.

- **Purpose**: The `Dense` layer represents a fully connected layer in a neural network where each neuron computes a weighted sum of the inputs followed by an activation function.

- **Concepts:**
  - **Fully Connected Layer**: A layer where each neuron receives input from all neurons in the previous layer. It applies a linear transformation followed by a non-linear activation function.
  - **Mathematical Operation**: In a `Dense` layer, the output is computed as a weighted sum of the inputs plus a bias term, which is then transformed by an activation function.

**Concepts:**

1. **`Sequential` Model**:

- **Layer-by-Layer Construction**: Layers are added in a specific order, and the output of one layer is directly connected to the input of the next layer.
- **No Shared Connections**: Unlike more complex models like functional APIs or subclassing, `Sequential` models do not allow for multiple inputs or outputs or shared layers between different parts of the model.

### **1. `optimizer='adam'`**

#### **Theory**

- **Optimizer**: An algorithm or method used to change the attributes of the neural network such as weights and learning rate in order to reduce the losses.
- **Adam (Adaptive Moment Estimation)**: A popular optimization algorithm that combines the advantages of two other extensions of stochastic gradient descent, namely AdaGrad and RMSProp.

**Key Concepts:**
- **Adaptive Learning Rate**: Adam adjusts the learning rate for each parameter individually.
- **Moment Estimates**: Uses estimates of first (mean) and second moments (variance) of the gradients to adapt the learning rate.

**Why Adam?**
- **Efficiency**: Adam is computationally efficient.
- **Memory Requirements**: It has low memory requirements.
- **Performance**: It works well on large datasets and in the context of non-stationary objectives (where the optimal parameters change over time).

- **Loss Function**: A method to evaluate how well the algorithm models the given data. If predictions deviate from the actual results, the loss function will output a higher number.
- **Binary Crossentropy**: Also known as log loss, it is used as a loss function for binary classification problems.

**Key Concepts:**
- **Binary Classification**: Problems where there are two possible outcomes.
- **Cross-Entropy**: Measures the difference between two probability distributions for a given set of events or values.

**Why Binary Crossentropy?**
- **Probabilistic Interpretation**: It provides a probabilistic interpretation of the output, which is useful for binary classification.
- **Penalty for Wrong Predictions**: Gives a higher penalty to wrong predictions, encouraging the model to produce more accurate probabilities.

### Difference Between Cross-Entropy, Binary Cross-Entropy, and Categorical Cross-Entropy

**1. Cross-Entropy**

- **Description:** Cross-Entropy is a general concept used to measure the difference between two probability distributions: the true distribution (target) and the predicted distribution (model output).
- **Formula:** For two discrete probability distributions p (true distribution) and q (predicted distribution), cross-entropy is given by:
  `H(p, q) = -Σ p(x_i) * log(q(x_i))`
- **Use Case:** It is used in various classification problems to evaluate how well the model's predicted probability distribution matches the actual distribution.

**2. Binary Cross-Entropy**

- **Description:** Binary Cross-Entropy (also known as Log Loss) is a specific case of cross-entropy used for binary classification problems where there are only two classes (0 or 1).
- **Formula:** For binary classification with true labels y and predicted probabilities ŷ, the binary cross-entropy loss is:
  `Binary Cross-Entropy = - (1/N) * Σ [y_i * log(ŷ_i) + (1 - y_i) * log(1 - ŷ_i)]`
  Where y_i is the true label for the i-th sample, ŷ_i is the predicted probability for the positive class, and N is the number of samples.
- **Use Case:** Used for binary classification problems, such as classifying emails as spam or not spam.

**3. Categorical Cross-Entropy**

- **Description:** Categorical Cross-Entropy is used for multi-class classification problems where there are more than two classes.
- **Formula:** For multi-class classification with true labels y (one-hot encoded) and predicted probabilities ŷ, the categorical cross-entropy loss is:
  `Categorical Cross-Entropy = - (1/N) * Σ Σ y_{i,j} * log(ŷ_{i,j})`
  Where y_{i,j} is 1 if the i-th sample belongs to class j, and 0 otherwise; ŷ_{i,j} is the predicted probability for the i-th sample being in class j; and K is the number of classes.
- **Use Case:** Used for multi-class classification problems, such as classifying flowers into different species.

### When to Use Which Loss Function?

- **Binary Cross-Entropy:** Use this for binary classification problems where you have one output neuron with a sigmoid activation function.
- **Categorical Cross-Entropy:** Use this for multi-class classification problems where you have multiple output neurons with a softmax activation function.
- **Cross-Entropy:** This is a more general term and not directly used in code but refers to the concepts of binary and categorical cross-entropy.


# Regression Code

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
california = fetch_california_housing()
data = california.data
targets = california.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, targets, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
```

```python
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
history = model.fit(X_train, y_train, epochs=100, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
```

```python
loss, mae = model.evaluate(X_test, y_test)
print(f'Test Mean Absolute Error: {mae}')
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Mean Absolute Error over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error')
plt.legend()

plt.show()
```
