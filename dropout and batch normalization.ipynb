```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
import matplotlib.pyplot as plt

# Load and preprocess the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

# Define the neural network model with dropout
model_with_dropout = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),  # Dropout layer with probability 0.5
    layers.Dense(10)
])

# Compile the model
model_with_dropout.compile(optimizer='adam',
                           loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                           metrics=['accuracy'])

# Train the model with dropout
history_with_dropout = model_with_dropout.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

# Define the same model without dropout
model_without_dropout = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(512, activation='relu'),
    layers.Dense(10)
])

# Compile the model without dropout
model_without_dropout.compile(optimizer='adam',
                              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                              metrics=['accuracy'])

# Train the model without dropout
history_without_dropout = model_without_dropout.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

# Plot training loss for comparison
plt.plot(history_with_dropout.history['loss'], label='With Dropout')
plt.plot(history_without_dropout.history['loss'], label='Without Dropout')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
import matplotlib.pyplot as plt

# Load and preprocess the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

# Define the neural network model with batch normalization
model_with_batch_norm = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(512),
    layers.BatchNormalization(),  # Batch normalization layer
    layers.Activation('relu'),
    layers.Dense(10)
])

# Compile the model
model_with_batch_norm.compile(optimizer='adam',
                              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                              metrics=['accuracy'])

# Train the model with batch normalization
history_with_batch_norm = model_with_batch_norm.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

# Define the same model without batch normalization
model_without_batch_norm = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(512),
    layers.Activation('relu'),
    layers.Dense(10)
])

# Compile the model without batch normalization
model_without_batch_norm.compile(optimizer='adam',
                                 loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                                 metrics=['accuracy'])

# Train the model without batch normalization
history_without_batch_norm = model_without_batch_norm.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

# Plot training loss for comparison
plt.plot(history_with_batch_norm.history['loss'], label='With Batch Norm')
plt.plot(history_without_batch_norm.history['loss'], label='Without Batch Norm')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend()
plt.show()
```
