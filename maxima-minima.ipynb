```python
import numpy as np
import matplotlib.pyplot as plt

# Function definitions
def f1(x):
    return 2 * x**3 - 3 * x**2 + 6

def f2(x):
    return -3 * x**2 + 4 * x + 7

# First function (Example 1)
x_values = np.linspace(-2, 2, 400)
y_values = f1(x_values)

# Plotting the function
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='y = 2x^3 - 3x^2 + 6')
plt.title('Function: y = 2x^3 - 3x^2 + 6')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Finding critical points
critical_points_x = [0, 1]
critical_points_y = [f1(x) for x in critical_points_x]
plt.scatter(critical_points_x, critical_points_y, color='red', label='Critical Points')

# Annotation for critical points
for i, txt in enumerate(critical_points_y):
    plt.annotate(f'({critical_points_x[i]}, {round(txt, 2)})', (critical_points_x[i], txt), textcoords="offset points", xytext=(-15,-10), ha='center')

plt.legend()

plt.show()


# Second function (Example 2)
x_values = np.linspace(-2, 3, 400)
y_values = f2(x_values)

# Plotting the function
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label='f(x) = -3x^2 + 4x + 7')
plt.title('Function: f(x) = -3x^2 + 4x + 7')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Finding critical points
critical_points_x = [2/3]
critical_points_y = [f2(x) for x in critical_points_x]
plt.scatter(critical_points_x, critical_points_y, color='red', label='Critical Points')

# Annotation for critical points
for i, txt in enumerate(critical_points_y):
    plt.annotate(f'({round(critical_points_x[i], 2)}, {round(txt, 2)})', (critical_points_x[i], txt), textcoords="offset points", xytext=(-15,-10), ha='center')

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the symbolic variable
x = sp.Symbol('x')

# Define the function
f = 2*x**3 - 3*x**2 + 6

# Calculate the first derivative
f_prime = sp.diff(f, x)

# Calculate the critical points
critical_points = sp.solve(f_prime, x)

# Calculate the second derivative
f_double_prime = sp.diff(f_prime, x)

# Initialize lists to store maxima, minima, and saddle points
maxima = []
minima = []
saddle = []

# Iterate through critical points to determine maxima, minima, and saddle points
for point in critical_points:
    if f_double_prime.subs(x, point) > 0:
        minima.append(point)
    elif f_double_prime.subs(x, point) < 0:
        maxima.append(point)
    else:
        saddle.append(point)

# Define the function for plotting
f_plot = sp.lambdify(x, f, 'numpy')

# Generate x values for plotting
x_values = np.linspace(-2, 2, 400)

# Plot the function
plt.plot(x_values, f_plot(x_values), label='f(x) = 2x^3 - 3x^2 + 6')

# Plot maxima, minima, and saddle points
plt.scatter(maxima, [f.subs(x, point) for point in maxima], color='red', label='Maxima')
plt.scatter(minima, [f.subs(x, point) for point in minima], color='green', label='Minima')
plt.scatter(saddle, [f.subs(x, point) for point in saddle], color='blue', label='Saddle')

# Annotate points
for point in maxima:
    plt.annotate(f'Maxima ({point}, {f.subs(x, point)})', (point, f.subs(x, point)), textcoords="offset points", xytext=(-15,-10), ha='center')
for point in minima:
    plt.annotate(f'Minima ({point}, {f.subs(x, point)})', (point, f.subs(x, point)), textcoords="offset points", xytext=(-15,10), ha='center')
for point in saddle:
    plt.annotate(f'Saddle ({point}, {f.subs(x, point)})', (point, f.subs(x, point)), textcoords="offset points", xytext=(-15,10), ha='center')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Maxima, Minima, and Saddle Points')
plt.legend()

# Show plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
